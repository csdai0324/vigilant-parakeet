```python
import os
import re
import glob

import pandas as pd

from collections import Counter
```

```python
data_root = '/root/Medical_DataSet{}'
objs = glob.glob(data_root.format('/*'))
reports_csv_path = None
images_path = None
projections_csv_path = None
for obj in objs:
    if 'reports' in obj:
        reports_csv_path = obj
    elif 'proj' in obj:
        projections_csv_path = obj
    elif 'image' in obj:
        images_path = obj
objs
```

```python
df = pd.read_csv(reports_csv_path)
df = df.fillna('None.')
df.head(10)
```

```python
MeSH, findings, impression = df['MeSH'], df['findings'], df['impression']
```

```python
MeSH_counter = Counter()
for mesh in MeSH.to_list():
    mesh = re.split('\W+', mesh)
    MeSH_counter += Counter(mesh)
print(MeSH_counter.most_common())
```

```python
findings_counter = Counter()
for finding in findings.to_list():
    
    finding = finding.split(' ')
    findings_counter += Counter(finding)
findings_counter.most_common()
```
